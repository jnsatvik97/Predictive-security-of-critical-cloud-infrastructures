```python
# action is an integer that is either 0 or 1
# 0 denotes remaining in the same state. 1 denotes jumping to a different state 
# state is a numpy array of size 7 cooresponding to all the 7 features
# function returns an integer which denotes reward
def RewardFunction(state,action):
    f1=state[0]
    f2=state[1]
    f3=state[2]
    f4=state[3]
    f5=state[4]
    f6=state[5]
    f7=state[6]
    R1=0;
    R2=0;
    R3=0;
    R4=0;
    R5=0;
    R6=0;
    R7=0;
    
    if(f1==4 or f1==5):
        R1=1
    if(f2==3 or f2==4):
        R2=1
    if(f3==4 or f3==5):
        R3=1    
    if(f4==3 or f4==4):
        R4=1
    if(f5==3 or f5==4):
        R5=1    
    if(f6==3 or f6==4):
        R6=1
    if(f7==5 or f7==6 or f7==7 or f7==8):
        R7=1
    
    w1= 1000
    w2= 1000
    w3= 1000
    w6= 1000
    
    w4= 2000
    w5= 2000
    
    w7= 3000
    
    RiskMetric=w1*R1+w2*R2+w3*R3+w4*R4+w5*R5+w6*R6+w7*R7
    
    RiskThreshold=(w1+w2+w3+w4+w5+w6+w7)/2
    
    wa=1500;
    Ra=1
    
    if(action==0):
        if(RiskMetric<RiskThreshold):
            Ra=-1
            
    if(action==1):
        if(RiskMetric>RiskThreshold):
            Ra=-1
            
    return RiskMetric+(wa*Ra)        
    
    
    
    
    
```

```python
# action is an integer that is either 0 or 1
# State1/State2 is a numpy array of size 7 cooresponding to all the 7 features
# function returns an integer which denotes probability of reaching from state 1 to state 2 by performing action a 
def TransitionFunction(State1,action,State2):
    countTransition=0
    countTotal=0
    probability=0
    
    for i in range(0,279):
        if((State1==StateTransitionMatrix[i]).all()):
            countTotal=countTotal+1
            if((StateTransitionMatrix[i+1]==State2).all()):
                countTransition=countTransition+1
    
    if(countTotal is not 0):
        probability=countTransition/countTotal
        
    if(action==1):
        return probability
    if(action==0):
        if(State2==State1):
            return 0.8
        else:
            return 0.2*probability
```

```python
def RiskMetric(state):
    f1=state[0]
    f2=state[1]
    f3=state[2]
    f4=state[3]
    f5=state[4]
    f6=state[5]
    f7=state[6]
    R1=0;
    R2=0;
    R3=0;
    R4=0;
    R5=0;
    R6=0;
    R7=0;
    
    if(f1==4 or f1==5):
        R1=1
    if(f2==3 or f2==4):
        R2=1
    if(f3==4 or f3==5):
        R3=1    
    if(f4==3 or f4==4):
        R4=1
    if(f5==3 or f5==4):
        R5=1    
    if(f6==3 or f6==4):
        R6=1
    if(f7==5 or f7==6 or f7==7 or f7==8):
        R7=1
    
    w1= 1000
    w2= 1000
    w3= 1000
    w6= 1000
    
    w4= 2000
    w5= 2000
    
    w7= 3000
    
    RiskMetric=w1*R1+w2*R2+w3*R3+w4*R4+w5*R5+w6*R6+w7*R7
    return RiskMetric;
```
